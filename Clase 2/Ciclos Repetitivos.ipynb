### Tabla Amortizacion

```python
import pandas as pd
import plotly.express as px
import numpy as np
import math
```

```python
#D = float(input("Ingrese la deuda : " ))
#T = float(input("Ingrese la Tasa de Interes : "))
#N = int(input("Ingrese los periodos : "))

D = 1000000
T = 0.15
N = 10
```

```python
R = (D*T)/ (1 - pow((1+T),-N))
```

```python
R
```

```python
K= 0
```

```python
tabla_amortizacion = []
while K!=N:
    renglon = []
    I= D * T
    A = R - I
    D = D - A
    K += 1
    renglon.append(K)
    renglon.append(float(I))
    renglon.append(A)
    renglon.append(int(D))
    tabla_amortizacion.append(renglon)
```

```python
titulos = ['Periodo','Interes','Amortizacion','Deuda']
df_tabla_amortizacion = pd.DataFrame(tabla_amortizacion,columns= titulos)
```

```python
df_tabla_amortizacion
```

```python
fig = px.line(df_tabla_amortizacion, x='Periodo', y="Deuda")
fig.show()
```
